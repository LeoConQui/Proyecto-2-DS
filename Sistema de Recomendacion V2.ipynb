# Clase Track

Creamos una clase track que tiene como atributos los datos de cada cancion contenida en el dataset:
/*:
  - URI: es un id unico para cada cancion
  - Nombre
  - Artist info: el URI y nombre del artista
  - Parent playlist: la playlist a la que pertence la cancion

```python
class Track:
    
    # constructor de la clase cuyos parametros son self, track_dict y playlist
    def __init__(self, track_dict, playlist):
        # asignamos valores a los atributos de la clase
        self.uri = track_dict["track_uri"]
        self.name = track_dict["track_name"]
        self.artist_uri = track_dict["artist_uri"]
        self.artist_name = track_dict["artist_name"]
        self.playlist = playlist
    
    # to string    
    def __str__(self):
        return f"Track {self.uri} called {self.name} by {self.artist_uri} ({self.artist_name}) in playlist {self.playlist}."
  
    # metodo que "reproduce" una cancion
    def __repr__(self):
        return f"Track {self.uri}"
```

# Clase Playlist

Esta clase modela una playlist y tiene los siguientes atributos:

  - Nombre: el nombre de la playlist y si index -indice- asociado
  - Title: el titulo que tiene dicha playlist en el dataset de spotify
  - un diccionario con datos de la playlist
  - diccionario de las canciones
  - lista de los URI's de los artistas que estan en la playlist

```python
class Playlist: 
  
    # constructor de la clase con sus parametros  
    def __init__(self, json_data, index):
        self.name = f"playlist_{index}"
        self.title = json_data["name"]
        self.data = json_data
        self.tracks = {}
        self.artists = []
  
    # metodo que carga todas las canciones en la playlist a un json con data para la playlist
    def load_tracks(self):
        tracks_list = self.data["tracks"]
        self.tracks = {x["track_uri"] : Track(x, self.name) for x in tracks_list}
        self.artists = [x["artist_uri"] for x in tracks_list]
    
    # tostring
    def __str__(self):
        return f"Playlist {self.name} with {len(self.tracks)} tracks loaded."
  
    # metodo que "reproduce" la playlist
    def __repr__(self):
        return f"Playlist {self.name}"
```

# Clase JSON File

Esta clase maneja archivos JSON que contienen informacion sobre las playlists. Dichos archivos van a contener la siguiente informacion:

  - Nombre del archivo
  - indice para numerar las playlists
  - un diccionario con el archivo json para el archivo completo
  - diccionario de playlists (nombres de las playlists)


```python
import json
from os.path import join  # Import the join function from the os.path module

class JSONFile: 
    
    # constructor
    def __init__(self, data_path, file_name, start_index):
        self.file_name = file_name 
        self.start_index = start_index

        # abrimos el archivo json y lo cargamos al diccionario
        with open(join(data_path, file_name)) as json_file:
            json_data = json.load(json_file)
            self.data = json_data 
            # diccionario para guardar las playlists
            self.playlists = {}

    # metodo que carga todas las playlists al json de data
    def process_file(self):
        for i, playlist_json in enumerate(self.data["playlists"]):
            # creamos un objeto Playlist para cada playlist en el json de datos
            playlist = Playlist(playlist_json, self.start_index + i)
            # cargamos las canciones para la playlist
            playlist.load_tracks()
            # agregamos la playlist al diccionario de playlists
            self.playlists[playlist.name] = playlist

    # tostring
    def __str__(self):
        return f"JSON {self.file_name} has {len(self.playlists)} playlists loaded."
  
    
    def __repr__(self):
        return self.file_name
```
